```python
from openai import OpenAI
import config

YOUR_API_KEY = config.SONAR_API_KEY

messages = [
    {
        "role": "system",
        "content": (
            "You are an artificial intelligence assistant to first responders,"
            " and you need to engage in a helpful, detailed, polite conversation"
            "with users to help respond to disaster events detected from satellite"
            "imagery. Do not use resources in the future than the time period indicated by the user."
        ),
    },
    {   
        "role": "user",
        "content": (
            " Today is Jan 9th, and I just received a satellite image labeled as:"
             "a large plume of smoke rising from the ground from coordinates 34.01205° N, 118.18242° W"
             "Do you know of any more information in the area, and how we can best support in addressing"
             "this possible natural disaster, futher identifying the cause and "
        ),
    },
]

client = OpenAI(api_key=YOUR_API_KEY, base_url="https://api.perplexity.ai")

# chat completion without streaming
response = client.chat.completions.create(
    model="sonar-pro",
    messages=messages,
)
print(response)

# chat completion with streaming
response_stream = client.chat.completions.create(
    model="sonar-pro",
    messages=messages,
    stream=True,
)
for response in response_stream:
    print(response)
```
